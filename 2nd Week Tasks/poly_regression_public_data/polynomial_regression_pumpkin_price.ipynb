membuat model menggunakan polynomial regression

```python
#menyimpan data dari kolom "Number of Units" ke dalam variabel 'x'
x = df["Origin"]

#menyimpan data dari kolom "Manufacturing Cost" ke dalam variabel 'y'
y = df["High Price"]
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Mengimpor PolynomialFeatures dan LabelEncoder dari scikit-learn
from sklearn.preprocessing import PolynomialFeatures, LabelEncoder

# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Menggunakan LabelEncoder untuk mengubah kolom 'Origin' menjadi nilai numerik
df['Origin_Encoded'] = label_encoder.fit_transform(df['Origin'])

# Menggunakan LabelEncoder untuk mengubah kolom 'Variety' menjadi nilai numerik
df['Variety_Encoded'] = label_encoder.fit_transform(df['Variety'])

# Menggunakan LabelEncoder untuk mengubah kolom 'Item Size' menjadi nilai numerik
df['Item Size_Encoded'] = label_encoder.fit_transform(df['Item Size'])

# Menentukan kolom-kolom fitur numerik yang akan digunakan dalam analisis
numerical_features = ['Item Size_Encoded', 'Variety_Encoded', 'Origin_Encoded']
```

```python
# Menggabungkan kolom fitur numerik menjadi satu dataframe
X = df[numerical_features]

# Inisialisasi derajat polinomial yang akan diuji (hanya 4 derajat)
degrees = [1, 2, 3, 4]

# Membuat subplot 1x4 dalam sebuah figur dengan ukuran
fig, axes = plt.subplots(1, 4, figsize=(25, 6))

# Menyusun subplot dengan jarak antar subplot 3.0
fig.tight_layout(pad=3.0)

# Inisialisasi variabel untuk menyimpan nilai-nilai MSE
mse_values = []

# Iterasi melalui derajat polinomial yang berbeda (hanya 4 derajat)
for i, degree in enumerate(degrees):
    # Membuat objek PolynomialFeatures dengan derajat tertentu
    polynomial_features = PolynomialFeatures(degree=degree)

    # Mengubah fitur independen 'x' menjadi bentuk polinomial
    x_poly = polynomial_features.fit_transform(X)

    # Melatih model regresi linier pada fitur polinomial
    model = LinearRegression()
    model.fit(x_poly, y)

    # Memprediksi hasil
    y_pred = model.predict(x_poly)

    # Menghitung MSE (Mean Squared Error)
    mse = mean_squared_error(y, y_pred)
    mse_values.append(mse)

    # Membuat plot hasil polinomial regression
    ax = axes[i]
    ax.scatter(x, y, label='Data', color='blue')
    ax.plot(x, y_pred, color='red', label='Regresi Polinomial')
    ax.set_title(f'Degree {degree}\nMSE: {mse:.2f}')
    ax.legend()

# Menampilkan plot
plt.show()

# Menampilkan MSE untuk setiap derajat
for degree, mse in zip(degrees, mse_values):
    print(f'Degree {degree}: MSE = {mse:.2f}')
```
