```python
from google.colab import drive

#menentukan direktori tempat Google Drive akan di-mount yaitu '/drive'
drive.mount('/drive')

#menentukan lokasi file CSV yang akan digunakan
file_path = '/drive/MyDrive/Dataset/Social_Network_Ads (1).csv'
```

```python
import pandas as pd

#menyimpan file CSV ke dalam variabel 'dframe'
dframe = pd.read_csv(file_path)
```

```python
#menampilkan lima baris pertama dari dataframe
dframe.head()
```

```python
#menampilkan informasi rinci dari dataframe
dframe.info()
```

```python
#menghapus kolom 'User ID' dari dataframe 'dframe' karena tidak diperlukan dalam melatih model
dframe = dframe.drop(columns=['User ID'])
```

```python
#menggunakan One-hot encoding untuk konversi kolom-kolom kategorikal menjadi variabel dummy biner yang dapat digunakan dalam analisis
dframe = pd.get_dummies(dframe)
dframe
```

```python
#menentukan kolom-kolom yang akan digunakan sebagai fitur (X) dan target (y)
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']

#memilih kolom-kolom yang sesuai dari dataframe 'dframe' dan menyimpannya sebagai fitur (X) dan target (y)
X = dframe[predictions]
y = dframe['Purchased']
```

```python
#import StandardScaler dari sklearn.preprocessing untuk melakukan penskalaan standar pada fitur-fitur dalam dataframe X
from sklearn.preprocessing import StandardScaler

#membuat objek scaler dari StandardScaler
scaler = StandardScaler()

#menyesuaikan scaler dengan data fitur X untuk menghitung mean dan standar deviasi
scaler.fit(X)

#mengubah data fitur X ke dalam skala standar menggunakan fungsi transform
scaled_data = scaler.transform(X)

#konversi hasil penskalaan ke dalam dataframe dengan nama kolom yang sesuai
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)

#menampilkan lima baris pertama dari dataframe yang telah diubah
scaled_data.head()
```

```python
#import train_test_split dari sklearn.model_selection untuk membagi data menjadi dua set, yaitu data pelatihan (train) dan data pengujian (test)
from sklearn.model_selection import train_test_split

#membagi data menjadi data pelatihan (train) dan data pengujian (test)
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
#import kelas LogisticRegression dari sklearn.linear_model untuk melatih model logistic regression
from sklearn import linear_model

#membuat objek model logistic regression
model = linear_model.LogisticRegression()

#melatih model dengan data pelatihan
model.fit(X_train, y_train)
```

```python
#menghitung skor akurasi model
model.score(X_test, y_test)
```
